````python
start_date = '20200101'
end_date = '20200125'
call = cft.CfRpl(collection='chm_v1', product='o3', lat = 45.2, lon=-85, start_date = start_date, end_date = end_date)
data = call.get_data(return_as='dataframe')
data['day_name'] = data['time'].dt.day_name()
print(data)
````

````python
#hour_mean = data.groupby([data["time"].dt.hour])["O3"].mean()
#hour_std = data.groupby([data["time"].dt.hour])["O3"].std()
hour_mean = data.groupby([data["day_name"]])["O3"].mean()
hour_std = data.groupby([data["day_name"]])["O3"].std()
hour_low = hour_mean - hour_std
hour_high = hour_mean + hour_std
````

````python
import matplotlib.pyplot as plt
````

````python
fig, ax = plt.subplots()
ax.plot(hour_mean.index, hour_mean.values, c = 'red', ls = '-', label='test', zorder=2)
ax.fill_between(hour_mean.index, 
                y1 = hour_low.values, 
                y2 = hour_high.values, 
                color = (1,0,0,0.5), 
                zorder=1)
````
